## Perform Exploratory Data Analysis (EDA)

### Variable Identification

```python
data.dtypes
```

* The *pickup_datetime* and *dropoff_datetime* variables needs to be changed into datetime.
* The *store_and_fwd_flag* variable needs to be changed into category.

```python
data['pickup_datetime'] = data['pickup_datetime'].astype('datetime64[ns]')
data['dropoff_datetime'] = data['dropoff_datetime'].astype('datetime64[ns]')

data['store_and_fwd_flag'] = data['store_and_fwd_flag'].astype('category')
```

### Checking Null and Unique values

```python
# Check if we have null values in the dataset

data.isnull().sum()
```

We don't have any null values in the dataset.

```python
# Check how many unique values are there in each column

data.nunique()
```

* There are 729322 rows in *id* column which is equal to the total number of rows in our dataset. Therefore, every row has unique *id*.
* There are 2 unique values for *Vendor_id* and *store_and_fwd_flag* columns which is as per the column description.

### Statistical values for each Numerical variable

```python
data.describe()
```

* The count of all the variables are equal to the total number of rows meaning we have no null values. Also, this was seen earlier as well. 
* The maximum value of *vendor_id* is 2 which we have seen above as well.
* The minimum value of *passenger_count* field is 0 which we need to undertand the reason. The maximum value is 9.
* The longitude and latitude values are not giving much insights right now.
* The average *trip_duration* is 15 minutes with minimum and maximum duration of 0.01 minute (We need to dig further to understand the reason) and 538 hours.

### New Feature Creation

```python
# Extracting day name
data['pickup_day'] = data['pickup_datetime'].dt.day_name()
data['dropoff_day'] = data['dropoff_datetime'].dt.day_name()

# Extracting week number
data['pickup_day_no'] = data['pickup_datetime'].dt.weekday
data['dropoff_day_no'] = data['dropoff_datetime'].dt.weekday

# Extracting month name
data['pickup_month_name'] = data['pickup_datetime'].dt.month_name()
data['dropoff_month_name'] = data['dropoff_datetime'].dt.month_name()

data['pickup_month_no'] = data['pickup_datetime'].dt.month
data['dropoff_month_no'] = data['dropoff_datetime'].dt.month

# Extracting hour from time
data['pickup_hour'] = data['pickup_datetime'].dt.hour
data['dropoff_hour'] = data['dropoff_datetime'].dt.hour
```

```python
def cal_distance(pickup_lat,pickup_long,dropoff_lat,dropoff_long):
    start_coordinates=(pickup_lat,pickup_long)
    stop_coordinates=(dropoff_lat,dropoff_long)
 
    return great_circle(start_coordinates,stop_coordinates).km
```

```python
data['distance'] = data.apply(lambda x: cal_distance(x['pickup_latitude'], x['pickup_longitude'], x['dropoff_latitude'], x['dropoff_longitude']), axis = 1)
```

### Pickup Days and Drop off Days - Catgorical Nominal Variables

#### Month wise Count

```python
figure,(ax1,ax2) = plt.subplots(ncols = 2,figsize = (20,5))
ax1.set_title('Pickup Month')
ax = sns.countplot(x = 'pickup_month_no',data = data,ax = ax1)

ax2.set_title('Dropoff Month')
ax = sns.countplot(x = 'dropoff_month_no', data = data, ax = ax2)
```

* Mostly all the months are busy but *March* and *April* are the busiest.

#### Day wise Count

```python
figure,(ax1,ax2) = plt.subplots(ncols = 2,figsize = (20,5))
order = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]

ax1.set_title('Pickup Days')
ax = sns.countplot(x = 'pickup_day',data = data,ax = ax1, order = order)

ax2.set_title('Dropoff Days')
ax = sns.countplot(x = 'dropoff_day', data = data, ax = ax2, order = order)
```

* Friday is the busiest days followed by Saturday.

#### Hour wise Count

```python
figure,(ax1,ax2) = plt.subplots(ncols = 2,figsize = (20,5))
ax1.set_title('Pickup Hour')
ax = sns.countplot(x = 'pickup_hour',data = data,ax = ax1)

ax2.set_title('Dropoff Hour')
ax = sns.countplot(x = 'dropoff_hour', data = data, ax = ax2)
```

* Both *pickup_hour* and *dropoff_hour* are similar.
* Peak time is during 6:00 PM to 7:00 PM.
* Non peak hour is during 3:00 AM to 5:00 AM.

#### Correlation Matrix

```python
numerical = data.select_dtypes(include=['int64','float64','Int64'])[:]
correlation = numerical.dropna().corr()
```

```python
# plotting heatmap usill all methods for all numerical variables
plt.figure(figsize=(36,6), dpi=140)
for j,i in enumerate(['pearson','kendall','spearman']):
  plt.subplot(1,3,j+1)
  correlation = numerical.dropna().corr(method=i)
  sns.heatmap(correlation, linewidth = 2)
  plt.title(i, fontsize=18)
```

* Kendall and Spearman correlation have very similar pattern except the difference between magnitude.
* Correlation lies between Drop off hour and Pickup hour.
